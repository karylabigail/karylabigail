The different classes for this dataset is about fetal health which are represented as 1 for Normal, 2 for Suspect, and 3 for Pathological. 
* Normal indicates that the fetal health parameters assessed through FHR monitoring fall within expected ranges, showing patterns and characteristics that are consistent with a healthy fetus.
* Suspect suggests that there are certain features or patterns observed in the FHR monitoring data that raise concerns or indicate potential issues with fetal well-being.
* Pathological indicates significant abnormalities or deviations from normal FHR patterns that are indicative of fetal distress.

Specifically, this will be very useful within the field of obstetrics and gynecology, specifically in prenatal and fetal monitoring. The development of machine learning models based on this dataset can offer valuable assistance to OB-GYNs and midwives in assessing fetal health during labor by predicting the likelihood of various health complications. These models predict the likelihood of various health complications, enhancing the evaluation of maternal-fetal well-being.

In practice, if a model predicts a "Suspect" classification, healthcare providers can proactively recommend additional interventions, such as increased monitoring, diagnostic tests, or consultations with specialists. This proactive approach facilitates the early identification of potential fetal complications, allowing more time for preventative measures and leading to better outcomes for both mother and baby.

SisPorto, developed by Omniview, is already in use and is a real life use case for this dataset. It operates as a software integrated into maternity rooms, concurrently providing real-time alerts. Its implementation demonstrates how machine learning models can directly impact patient care. **Description of features in the data set**
1. **baseline value** - This is the baseline fetal heart rate (FHR). Normal fetal heart rate ranges from 110 to 160 beats per minute (bpm)
2. **acceleration** - This is the number of accelerations of the FHR per second. They are short term rises in FHR of at least 15 bpm.
3. **fetal_movement** - This is the number of fetal movements per second.
4. **uterine_contractions** - This is the number of uterine contractions per second.
5. **light_decelerations** - This is the number of light decelerations per second. Light decelerations are characterized by a gradual decrease in the fetal heart rate.
6. **severe_decelerations** - This is the numner of severe decelerations per second. Severe decelerations are characterized by a more pronounced and abrupt decrease in the fetal heart rate.
7. **prolonged_decelerations** - This is abnormal FHR deceleration lasting from 2 to 10 minutes.
8. **percentage_of_time_with_abnormal_short_term_variability** - This is the percentage of time with abnormal short-term variability. They are irregular fluctiations in FHR measured in 1-minute intervals. This may be indicative of acute factors, such as changes in oxygenation levels, fetal movement and contractions.
9. **mean_value_of_short_term_variability** - This is the mean value of short-term variability. This is how much FHR changes in average within a specified period.
10. **percentage_of_time_with_abnormal_long_term_variability** - This is the percentage of time with abnormal long-term variability. They are irregular fluctuations in FHR over extended periods of time (hours or days). They can be indicative of chronic factors such as oxygen deprivation, maternal health, and placental insufficiency.
11. **mean_value_of_long_term_variability** - This is the mean value of long-term variability. This is the overall value of patterns of fluctuations and trends in FHR over an extended period of time.
12. **histogram_width** - This represents the spread or width of the distribution of the FHR values. A wider histogram width has greater variability in FHR, which is considered a good indicator. On the other hand, a narrower histogram width has less variability in FHR.
13. **histogram_min** - This is the lowest value recorded within the FHR histogram.
14. **histogram_max** - This is the highest value recorded within the FHR histogram.
15. **histogram_number_of_peaks** - This is the count of distinct peaks or spikes. Higher number of peaks indicate fluctuations while lower number of peaks indicate more stable FHR patterns.
16. **histogram_number_of_zeroes** - This is the count of intervals or bins in the histogram where no FHR values are recorded.
17. **histogram_mode** - This is the most frequently recurring FHR value.
18. **histogram_mean** - This is the average FHR value recorded.
19. **histogram_median** - This is the middle value of the FHR data.
20. **histogram_variance** - This is the measure of the dispersion or spread of FHR values around the mean value.
21. **histogram_tendency** - This was a one-hot encoding where 1 means right asymmetric tendency, -1 means left asymmetric tendency and 0 means symmetric tendency.
